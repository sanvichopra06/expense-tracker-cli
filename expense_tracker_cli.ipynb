# Expense Tracker CLI Project

A beginner Python project using lists, loops, and conditionals to track daily expenses.

## Features

- Add Expense
- Show Expenses
- Show Total Expense
- Show Maximum Expense
- Show Minimum Expense
- Delete an Expense
- Exit Program

```python
def add_expense(expenses):
    exp= float(input("Add Expense: "))
    expenses.append(exp)
    print("Added Successfully")
    print()
def show_expenses(expenses):
    print("Expenses: " ,expenses, end=" ")
    print()
def total_expense(expenses):
    print("Total Expense: ",sum(expenses))
    print()
def max_expense(expenses):
    if len(expenses) == 0:
        print("No Expense")
    else:
        print("Maximum Expense: ", max(expenses))
        print()
def min_expense(expenses):
     if len(expenses) == 0:
        print("No Expense")
     else:
        print("Minimum Expense: ", min(expenses))
        print()
def delete_expense(expenses):
    target= float(input("Enter Expense to Delete: "))
    flag= False
    for i in range(len(expenses)):
        if expenses[i]==target:
            flag= True
            expenses.remove(expenses[i])
            print("Expense Deleted")
            break
    if flag== False:
        print("Expense Not Found")
```

```python
expenses =[] 
while True:
    print("1. Add Expense \n2. Show Expenses\n3. Show Total Expenses\n4. Maximum Expense\n5. Minimum Expense\n")
    print("6. Delete Expense\n7. Exit")
    choice= input("Enter your Choice: ")
    print()
    if choice == '1': 
        add_expense(expenses)
    elif choice == '2': 
        show_expenses(expenses)
    elif choice == '3': 
        total_expense(expenses)
    elif choice == '4':
        max_expense(expenses)
    elif choice == '5':
        min_expense(expenses)
    elif choice == '6':
        delete_expense(expenses)
    elif choice == '7': 
        print("Exited")
        break
    else: print("Invalid Choice")
    print()
```
